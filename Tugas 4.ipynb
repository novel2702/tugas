```python


# jumlah pengeluaran apbn

data_apbn.nilai.sum()

```

```python


# median pada kolom nilai apbn

data_apbn.nilai.median()

```

```python
# standar deviasi pada kolom nilai apbn

round (data_apbn.nilai.std(), 2)
```

```python

# rata-rata nilai apbn menurut tipe pengeluaran

data_apbn.groupby('tipe_pengeluaran').sum()['nilai']

```

```python
# rata-rata nilai apbn menurut pos belanja

data_apbn.groupby('pos_belanja').sum()['nilai']
```

```python
# jumlah guru berdasarkan tingkat pendidikan dan kelompok usia

data_apbn.groupby(['tahun','tipe_pengeluaran']).sum()['nilai']
```

```python
ax = data_apbn.groupby(['tipe_pengeluaran']).sum()['nilai'].T
ax = ax.plot(kind='bar', title="Rata-Rata pengeluaran apbn", legend=True, fontsize=14, figsize=(18,6))
ax.set_xlabel('Tipe pengeluaran', fontsize=14)
ax.set_ylabel('Nilai', fontsize=14)
```

```python


ax = data_apbn.groupby('tahun').sum()['nilai']
ax = ax.plot(kind='bar', title="Pengeluaran APBN berdasarkan tahun", legend=True, fontsize=12)
ax.set_xlabel('Tahun', fontsize=14)
ax.set_ylabel('Nilai', fontsize=14)

```
